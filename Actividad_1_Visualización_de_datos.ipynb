```python
# Generando puntos de una funcion
x_val = np.linspace (0,10,20)
y_val = x_val ** 2
plt.scatter(x_val, y_val)
```

HISTOGRAMA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
# Uso de funci√≥n multivariate_normal de la biblioteca NumPy para generar datos aleatorios de una distribuci√≥n normal multivariada
# argumentos media [0,0], matriz de covarianza [[5,2],[2,2]], tamano de muestra size 2000
data = pd.DataFrame(data, columns=['x', 'y'])
for col in 'xy':
      plt.hist(data[col], alpha=0.5)
```

La grafica anterios evidencia que los datos de 'x' y 'y' presentan una distribucion normal, se observa que valores positivos y negativos son igualmente probables, ya que la distribucion es simetrica respecto a la media cero, ademas la distribucion presenta forma gausiana, por tanto la distribuci√≥n es normal. La distribuci√≥n de datos respresentados por el color rojo presentan una densidad levemente mayor, que la distribuci√≥n de datos en color rojo, ya que los datos de la primera est√°n mas concentrados respecto a la media, que se ubica al rededor de cero

```python
for col in 'xy':
  sns.kdeplot(data[col], shade=True)
```

```python
for col in 'xy':
  sns.kdeplot(data[col], fill=True)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y'])
```

En esta gr√°fica con dos distribuciones se observa par ambas una distribucion normal o gausiana, explicada por la simetria de distribucion de los datos con respecto a la media, con la diferencia que una de ellas (curva color naranja) presenta un valor mas concentrado en comparaci√≥n a la distribuci√≥n en coloraci√≥n azul muestra una dispersion de datos levemente mayor con respecto a la media, evidenciado en la anchura de la campana. Ambas tienen una media al rededor de cero. La curva naranja presenta una distribuci√≥n estandar menor porque no se va hacia los extremos, lo que indica una densidad mayor que la distribuci√≥n identificada con el color azul, que tiende a prolongarse hacia los extremos, lo que indica una distribuci√≥n estandar mas grande, por tanto la distribuci√≥n en torno a la media no es tan estrecha si se compara con la curva de color naranja

```python
sns.displot(data['x'])
sns.displot(data['y'])
```

```python
sns.histplot(data['x'])
sns.histplot(data['y'])
```

```python
iris = sns.load_dataset("iris")
iris.head()
sns.pairplot(iris, hue='species', size=2.5);
```

Con los datos de la figura anterior es posible mostrar que el ancho de p√©talos permite separar las especies f√°cilmente, tambi√©n es posible ver que esta situaci√≥n se parece al largo del p√©talo. Sin embargo se ve que con el ancho del s√©palo no es f√°cil dividir las flores por tipo, asi mismo el largo del sepalo tampoco permite clasificar estas especies con facilidad.

Al analizar la caracter√≠stica referente al ancho de los p√©talos se observa que la flor setosa presenta una menor disperci√≥n de los datos respecto de la media cuyo valor aproximado es 0.3 aproximadamente, por tanto mayor densidad en comparaci√≥n a las gr√°ficas de distribuci√≥n de la flores versicolor y virginica
